```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
from sklearn import preprocessing
%matplotlib inline
```

```python
df = pd.read_csv("ClusterPlot.csv", usecols=[1,2])
df

plt.scatter(df['V1'], df['V2'])
plt.show()
```

```python
x=df.copy()
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
cluster=x.copy()
cluster['cluster_pred']=kmeans.fit_predict(x)
```

```python
plt.scatter(cluster['V1'], cluster['V2'], c=cluster['cluster_pred'], cmap='rainbow')
plt.show()
```

```python
x_scaled= preprocessing.scale(x)
x_scaled
```

```python
wcss=[]

for i in range(1,30):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,30),wcss)
plt.show()
```

```python
kmeans_new=KMeans(4)
kmeans.fit(x_scaled)
cluster_new=x.copy()
cluster_new['cluster_pred']=kmeans_new.fit_predict(x_scaled)
cluster_new
```

```python
plt.scatter(cluster_new['V1'], cluster_new['V2'], c=cluster_new['cluster_pred'], cmap='rainbow')
plt.show()
```
